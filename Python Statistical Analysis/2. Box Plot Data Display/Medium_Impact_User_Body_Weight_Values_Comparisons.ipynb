Box Plot showing the Values for the *'Medium Impact User Body Weight (kg)'* for the first defined stiffness category for each of the manufacturer companies. This is one of the variables that allows for the comparison of stiffness categories. The boxplot will show the distrubtuion of data values, allowing the standard error to be calculated.


```python
import matplotlib.pyplot as plt
import numpy as np

# data set values of the 'Medium Impact User Body Weight (kg)'
data1 = [52.2, 57.2, 64, 59.9, 68.4]

# calculate the Mean and Standard Deviation
mean_value1 = np.mean(data1)
std_deviation1 = np.std(data1)

# display the mean value and the standard deviation
plt.text(1, mean_value1, f'Mean: {mean_value1:.2f}', verticalalignment='bottom', horizontalalignment='right', color='green')
plt.text(1, std_deviation1, f'Standard Deviation: {std_deviation1:.2f}', verticalalignment='bottom', horizontalalignment='right', color='blue')

# create a boxplot
plt.boxplot(data1)

# add labels and a title to the boxplot
plt.xlabel('Stiffness Category 1, Data Set 1')
plt.ylabel('Specified Weight (kg)')
plt.title('Medium Impact User Body Weight (kg) values for each second provided stiffness category')

plt.show()
```
